```python
# Exploratory Data Analysis

import numpy as np
import pandas as pd
```

```python
# Load the Dataset

df = pd.read_csv('https://trello-attachments.s3.amazonaws.com/600d1f10d700af20b1924b3c/600d1f700fcd073384905133/4875f4c4af388917d8508e224b3d7219/telecom_churn.csv')
df.head()
```

```python
# columns  -->  to retrieve the column names
df.columns
```

```python
# check the dimensions

df.shape
```

```python
# info()  -->  to generate the full summary of dataset
df.info()
```

```python
# changing the datatype a column to diff datatype

# astype()  -->  to change the datatype
df['churn'] = df['churn'].astype('int64')
df.info()
```

```python
# describe()
# to give statistical summary of dataset
# by default, this method works only on integer and float values

df.describe()
```

```python
# in order to see the statistical summary of non numerical features,use include parameter in describe()

df.describe(include=['object','bool'])
```

```python
data = pd.Series(['A','B','C','A','C','D','E','C','B','C','B','A'])
data.describe()

# top  -->  most repeated characters : C
# freq  -->  amount of repititions : 4
```

```python
# value_counts()
df['churn'].value_counts()

# 0 --> active users
# 1 --> churn users  (stop subscriptions)
```

```python
df.head(3)
```

```python
# sorting


df.sort_values(by='total day charge').head(10)   # ascending order, by DEFAULT
```

```python
# descending order
df.sort_values(by='total day charge',ascending=False).head(10)
```

```python
df.sort_values(by=['total day minutes','total day calls']).head(10)     # to sort more than 1  (but commonly cause errors)
```

```python
# maximum value of total day charge
df['total day charge'].max()
```

```python
# how much time on an average do churn users spend on phone during daytime?

df[df['churn']==1]['total day minutes'].mean()
```

```python
df[df['churn']==0]['total day minutes'].mean()    # active users
```

```python
# what is the maximum length of international calls among loyal users?

df[df['churn']==0]['total intl calls'].max()
```

```python
df.max()
```

```python
# retrieve only those records where state starts with the letter 'w'
df[df['state'].apply(lambda state: state[0]=='W')].head()
```
