```python
# Connecting to a Google

import pandas as pd
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['Covid 19 prevention','prevention of Covid 19'])
```

```python
# Interest by region()

data = pytrend.interest_by_region()  # fetch the worldwide information

data
```

```python
pd.set_option("display.max_rows",None,"display.max_columns",None)
```

```python
data
```

```python
# Plot the result on Bar Graph

data.reset_index().plot(x='geoName',y='Covid 19 prevention',figsize=(120,10),kind='bar')
```

```python
kw_list = ['prevention of covid 19','prevention for covid 19','covid-19 prevention']
pytrend.build_payload(kw_list , timeframe='2021-01-01 2021-06-14',geo='MY')

# timeframe  --> Date_To_Start_From   End_Date  YYYY-MM-DD
# geo --> Two Letters Country Abbreviation
```

```python
malaysia_data = pytrend.interest_over_time()
malaysia_data
```

```python
data = malaysia_data.drop(labels=['isPartial'],axis='columns')
image = data.plot(title='Covid 19 Prevention Analysis On Malaysia')
fig = image.get_figure()
fig.set_size_inches(18.5,10.5)
```

```python
# Saving the result in CSV File

data.to_csv('search_trends.csv')
```

```python
import requests

# Getting The Data

url_request = requests.get('https://services1.arcgis.com/0MSEUqKaxRlEPj5g/arcgis/rest/services/Coronavirus_2019_nCoV_Cases/FeatureServer/1/query?where=1%3D1&outFields=*&outSR=4326&f=json')
url_json = url_request.json()
data = pd.DataFrame(url_json['features'])
```

```python
newData = data['attributes'].tolist()
df = pd.DataFrame(newData)
df
```

```python
df.set_index('OBJECTID')
```

```python
df.columns
```

```python
df = df[['Province_State', 'Country_Region','Confirmed', 'Recovered', 'Deaths', 'Active']]
df
```

```python
df.columns = ('State', 'Country','Confirmed', 'Recovered', 'Deaths', 'Active')
```

```python
df.head()
```

```python
import plotly.express as px
top10_confirmed = pd.DataFrame(df.groupby('Country')['Confirmed'].sum().nlargest(10).sort_values(ascending=False))
```

```python
fig1 = px.scatter(top10_confirmed, x=top10_confirmed.index , y='Confirmed',size='Confirmed',
                  title='Top 10 Countries')
fig1.show()
```
